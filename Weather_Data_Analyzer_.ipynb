```python
 region= input("Enter Region Name: ")
# Lists to store temperature and humidity values
temperature = []
humidity = []

days = int(input("Enter number of days of weather data: "))

# Input weather data
for i in range(days):
    temp = float(input(f"Enter temperature for day {i+1}: "))
    hum = float(input(f"Enter humidity for day {i+1}: "))

    temperature.append(temp)
    humidity.append(hum)

# average temperature
def average_temperature(temp_list):
    avg_temp = sum(temp_list) / len(temp_list)
    return avg_temp

#  average humidity
def average_humidity(hum_list):
    avg_hum = sum(hum_list) / len(hum_list)
    return avg_hum

#checking
def weather_condition(avg_temp, avg_hum):
    if avg_temp > 30 and avg_hum < 60:
        return "Sunny"
    elif avg_hum > 80:
        return "Rainy"
    elif avg_temp < 20:
        return "Cold/Cloudy"
    else:
        return "Moderate Weather"

# Calculate averages
avg_temp = average_temperature(temperature)
avg_hum = average_humidity(humidity)

# Determine condition
condition = weather_condition(avg_temp, avg_hum)

# Print results
print("\n----- Weather Data Analysis -----")
print("Region:", region)
print("Average Temperature:", round(avg_temp, 2), "°C")
print("Average Humidity:", round(avg_hum, 2), "%")
print("Weather Condition:", condition)
```
